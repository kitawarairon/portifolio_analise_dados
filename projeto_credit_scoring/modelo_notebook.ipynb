```python
import pandas as pd
import numpy as np
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from biokit.viz import corrplot

# configurações
import warnings
warnings.filterwarnings('ignore')
```

# Análise do dataset
---

```python
medidas = [[1700.26, 10.76],
[1746.49, 10.18],
[1804.37, 9.58],
[1878.73, 9.12],
[1931.44, 8.76],
[2014.44, 8.46],
[2086.20, 8.21]]
```

```python
frame = pd.DataFrame(medidas, columns=['Custo (R$)', 'h (m)'], index=[x for x in range(20, 55, 5)])
frame.index.name = 'fck'
frame
```

```python

data1 = frame['Custo (R$)']
data2 = frame['h (m)']

fig, ax1 = plt.subplots()

plt.title('Variação do custo e da altura', fontsize=18, color='navy')
color = 'tab:blue'
ax1.set_xlabel('$f_{ck}$')
ax1.set_ylabel('Custo da laje maciça (R$)', color=color, fontsize=16)
ax1.plot(data1, color=color, label='Custo')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:orange'
ax2.set_ylabel('Altura da laje (cm)', color=color, fontsize=16) 
ax2.plot(data2, color=color, label='Altura')
ax2.tick_params(axis='y', labelcolor=color)

plt.grid()
fig.tight_layout() 
plt.show()
```
