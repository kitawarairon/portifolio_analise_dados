# Criação das metodologias

## Criando a rotina para selecionar o insumo

```python
def criar_dataframe(caminho, chave_insumo, sheets):
    
    # criando dataframe vazio
    data = pd.DataFrame()
    
    # iniciando a rotina
    for i in range(sheets):
        
        # acessando o dataframe
        df = pd.read_excel(caminho, sheet_name=sheets)
        
        # localizando index do aço e do concreto
        insumo = df.iloc[chave_insumo, 2:23]
        
        # adicionando ao dataframe
        data = data.append(insumo)
        
        # contagem
        sheets -= 1
        
    return data
```

## Formatando o dataframe

```python
def formatar_dataframe(data, nomes, nome):
    # fazendo cópia do dataframe
    frame = data.copy()
    
    # renomenado e atualizando
    frame.rename(mapper=nomes, axis=1, inplace=True)
    frame.index = pd.date_range(start = '01-2007', periods=169, freq = '1M')
    
    # criando formatadores
    formatar_virgulas = lambda x: x.replace(',','.') if type(x) is str else x
    formatar_dois_pontos = lambda x: x.replace('..','.') if type(x) is str else x
    
    # aplicacando a formatação
    frame = frame.applymap(formatar_virgulas)
    frame = frame.applymap(formatar_dois_pontos)
    frame = frame.applymap(float)
    
    # criando a coluna da mpedia
    frame['media_brasil'] = frame.loc[:,:].mean(axis=1)
    
    #exportando arquivo
    frame.to_csv(f'../dados_exportados/{nome}.csv', encoding='utf-8', index=False)
    return frame
```

## Criando alguns dataframes

```python
%%time
concreto = criar_dataframe('../dados/tab_06.A.12_4.xlsx', 8, 169)
concreto_df = formatar_dataframe(concreto, mapper, 'concreto')
```

```python
%%time
aco = criar_dataframe('../dados/tab_06.A.12_4.xlsx', 7, 169)
aco_df = formatar_dataframe(aco, mapper, 'aco')
```

# Analisando a variação do aço e do concreto ao longo do tempo

## Gráfico da variação do custo do aço e do concreto

```python
plt.figure(figsize=(20, 12), dpi=400, facecolor='white')
plt.grid()
plt.title('Variação do custo do aço e do concreto ao longo dos anos', fontsize=24)


plt.subplot(2,1,1)
plt.plot(concreto_df.index, concreto_df.media_brasil, color='red', linestyle='--')
plt.title('Variação do custo do aço e do concreto ao longo dos anos', fontsize=24)
plt.xlabel('Tempo (meses)', fontsize=14)
plt.ylabel('Custo do $m^{3}$ do concreto', fontsize=16)
plt.grid()


plt.subplot(2,1,2)
plt.plot(aco_df.index, aco_df.media_brasil, color='green', linestyle='--')
plt.xlabel('Tempo (meses)', fontsize=14)
plt.ylabel('Custo do $kg$ do aço', fontsize=16)
plt.grid()
```

## Gráfico da média móvel agrupando os dois gráficos

```python
plt.figure(figsize=(18, 6), dpi=400, facecolor='white')
plt.grid()
plt.title('Variação do custo do aço e do concreto ao longo dos anos', fontsize=24)


plt.plot(concreto_df.index, concreto_df.media_brasil/concreto_df.media_brasil.iloc[0], color='red', linestyle='--', label='Concreto')
plt.plot(aco_df.index, aco_df.media_brasil/aco_df.media_brasil.iloc[0], color='green', linestyle='--', label='Aço')

plt.yticks([x/10 for x in range(10, 30, 1)])

plt.xlabel('Tempo (meses)', fontsize=14)
plt.ylabel('Custo normatizado', fontsize=18)

plt.legend()
```
