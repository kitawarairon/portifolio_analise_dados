# Clusterização

## Separando as variáveis

```python
X = wines
```

## Pré-processamento das variáveis

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
X_mms = mms.fit_transform(X)
```

## Métricas de avaliação

* [Silhoutte](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html)

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

* [Davies Bouldin](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.davies_bouldin_score.html)

The minimum score is zero, with lower values indicating better clustering.

* [Calinski harabasz](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.calinski_harabasz_score.html)

The score is higher when clusters are dense and well separated, which relates to a standard concept of a cluster.

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

def validar_cluster(values, labels):
    sil = silhouette_score(values, labels)
    db = davies_bouldin_score(values, labels)
    ch = calinski_harabasz_score(values, labels)
    validadores = {'Silhoutte':sil, 'Davies Bouldin':db, 'Calinski Harabasz':ch}
    frame = pd.DataFrame(validadores.values(), index=validadores.keys(), columns=['Values']).round(3)
    return frame

def metrics_values(values, labels):
    sil = silhouette_score(values, labels)
    db = davies_bouldin_score(values, labels)
    ch = calinski_harabasz_score(values, labels)
    return sil, db, ch
```

## KMeans

```python
from sklearn.cluster import KMeans
```

### Criando o modelo

```python
kmeans = KMeans(n_clusters=3, random_state=25)
kmeans.fit(X_mms)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=wines.Intensidade_Cor,
                         y = wines.Alcool, mode = 'markers',                      
                          marker = dict(color = kmeans.labels_.astype(np.float)),
                        text = kmeans.labels_))
             
```

```python
fig = go.Figure()

# criando o scatter
fig.add_trace(go.Scatter3d(
    
                        x=wines.Intensidade_Cor,
                        y = wines.Alcool,
                        z = wines.Linha_Pro,
                        mode = 'markers',                      
                        marker = dict(color = kmeans.labels_.astype(np.float)),
                        text = [0, 1, 2]
    
))

# renomeando os eixos
fig.update_layout(scene = dict(
                    xaxis_title='Intensidade de Cor',
                    yaxis_title='Àlcool',
                    zaxis_title='Proline'))


fig.show()
```

### Localizando os centroides

```python
centros = pd.DataFrame(kmeans.cluster_centers_)
centros.columns = wines.columns
centros.head()
```

### Plotando os centroides

```python

# criando dos centros
fig.add_trace(go.Scatter3d(
    
                        x=centros.Intensidade_Cor,
                        y = centros.Alcool,
                        z = centros.Linha_Pro,
                        mode = 'markers',                      
                        marker = dict(color = 'green'),
                        text = [0, 1, 2]
    
))

# renomeando os eixos
fig.update_layout(scene = dict(
                    xaxis_title='Intensidade de Cor',
                    yaxis_title='Àlcool',
                    zaxis_title='Proline'))


fig.show()
```

### Validação

**VALIDAÇÃO PELAS MÉTRICAS**

```python
validar_cluster(X_mms, kmeans.labels_ )
```

### Criando as categorias de vinhos

```python
catalogo_vinhos = wines.groupby(kmeans.labels_).mean()
catalogo_vinhos.index.name =  'VINHOS'
catalogo_vinhos.columns.name ='PROPRIEDADES'
catalogo_vinhos.index = ['VINHO A', 'VINHO B', 'VINHO C']
catalogo_vinhos = catalogo_vinhos.T
catalogo_vinhos['media'] = catalogo_vinhos.mean(axis=1)
catalogo_vinhos['var'] = catalogo_vinhos.var(axis=1)
catalogo_vinhos['std'] = catalogo_vinhos.std(axis=1)
catalogo_vinhos
```

## DBSCAN

**ATRIBUTOS:**

* **eps:** float, default=0.5

The maximum distance between two samples for one to be considered as in the neighborhood of the other. This is not a maximum bound on the distances of points within a cluster. This is the most important DBSCAN parameter to choose appropriately for your data set and distance function.

* **min_samples:** int, default=5

The number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself.

* **metric:** str, or callable, default=’euclidean’

The metric to use when calculating distance between instances in a feature array. If metric is a string or callable, it must be one of the options allowed by sklearn.metrics.pairwise_distances for its metric parameter. If metric is “precomputed”, X is assumed to be a distance matrix and must be square. X may be a Glossary, in which case only “nonzero” elements may be considered neighbors for DBSCAN.

Métricas de distância: [‘cityblock’, ‘cosine’, ‘euclidean’, ‘l1’, ‘l2’, ‘manhattan’, ‘braycurtis’, ‘canberra’, ‘chebyshev’, ‘correlation’, ‘dice’, ‘hamming’, ‘jaccard’, ‘kulsinski’, ‘mahalanobis’, ‘minkowski’, ‘rogerstanimoto’, ‘russellrao’, ‘seuclidean’, ‘sokalmichener’, ‘sokalsneath’, ‘sqeuclidean’, ‘yule’]

**Manuais:** 

* [DBSCAN - SKLEARN](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

* [DBSCAN](https://machinelearningknowledge.ai/tutorial-for-dbscan-clustering-in-python-sklearn/)


```python
from sklearn.cluster import DBSCAN
```

```python
%%time
dbscan = DBSCAN(eps=2, min_samples=15, metric='manhattan')
dbscan.fit(X_mms)
```

```python
metrics_values(X_mms, dbscan.labels_)
```

```python
dbscan.labels_
```

### Em busca das melhores métricas

```python
%%time
# gerando lista de testes
n = np.arange(10, 20, 5)

# criando listas auxiliares
sil = []
dbs = []
calinski = []

for i in list(n):
    print(i)
    model = DBSCAN(eps=2, min_samples=15, metric='manhattan')
    model.fit(X_mms)
    a, b, c = metrics_values(X_mms, model.labels_)   
    sil.append(a)
    dbs.append(b)
    calinski.append(c)
    
#criando dataframe    
frame = {
    'silhouette':sil,
    'davies_bouldin':dbs,
    'calinski':calinski,
        }

# ordenando o dataframe
tabela = pd.DataFrame(frame)
tabela.sort_values(by=['silhouette'], ascending=False).head()
```
